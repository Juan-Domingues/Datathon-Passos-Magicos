```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
dataset = pd.read_csv("../dataset_pede_tratado.csv")

print(dataset.shape)
```

```python
dataset[["RA","FASE"]].value_counts()
```

Pergunta 1 — Defasagem (IAN)

```python
plt.figure(figsize=(8,5))

sns.histplot(dataset["DEFAS"], bins=15)

plt.title("Distribuição da Defasagem dos Alunos")
plt.xlabel("Defasagem")
plt.show()

sns.boxplot(data=dataset, x="ANO", y="DEFAS")

plt.title("Evolução da Defasagem ao longo dos anos")
plt.show()

dataset["DEFAS"].value_counts(normalize=True)
```

A análise da defasagem educacional mostra que aproximadamente 79,6% dos alunos estão no nível esperado ou apresentam apenas leve defasagem em relação à fase ideal.

Cerca de 14% dos alunos apresentam defasagem moderada ou severa, configurando um grupo que pode demandar maior atenção pedagógica.

Apenas 6% dos alunos estão adiantados em relação à fase ideal, indicando que o programa atende majoritariamente alunos em situação de defasagem leve ou moderada.

A evolução ao longo dos anos mostra uma distribuição relativamente estável da defasagem, sugerindo que o programa contribui para manter os alunos próximos da progressão educacional esperada, evitando o agravamento da defasagem ao longo do tempo.

Pergunta 2 — Desempenho acadêmico (IDA)

```python
ida_ano = dataset.groupby("ANO")["IDA"].mean()

ida_ano.plot(marker="o")

plt.title("Evolução do IDA ao longo dos anos")
plt.ylabel("IDA médio")

plt.show()

sns.boxplot(data=dataset, x="FASE", y="IDA")
```

A análise do desempenho acadêmico médio (IDA) mostra uma melhora significativa entre 2022 e 2023, seguida por uma leve queda em 2024. Esse comportamento sugere que o desempenho acadêmico não apresenta uma tendência linear de crescimento, mas sim variações associadas possivelmente ao perfil dos alunos atendidos em cada ano.

Ao observar o desempenho ao longo das fases, percebe-se que as fases iniciais apresentam níveis mais elevados de IDA, seguidas por uma queda nas fases intermediárias e posterior recuperação nas fases mais avançadas. Esse padrão pode indicar que os alunos enfrentam desafios acadêmicos maiores durante a progressão educacional, mas conseguem recuperar desempenho à medida que se adaptam ao programa.

Além disso, as fases mais avançadas apresentam maior dispersão de desempenho, indicando maior heterogeneidade entre os alunos em níveis mais elevados do programa.

Pergunta 3 — Engajamento vs desempenho

```python
sns.scatterplot(data=dataset, x="IEG", y="IDA", alpha=0.4)

plt.title("Engajamento vs Desempenho Acadêmico")
plt.show()

sns.scatterplot(data=dataset, x="IEG", y="IPV", alpha=0.4)

plt.title("Engajamento vs Ponto de Virada")
plt.show()
```

A análise da relação entre engajamento nas atividades (IEG) e os indicadores de desempenho revela uma associação positiva consistente. Alunos com maiores níveis de engajamento tendem a apresentar melhores resultados tanto no desempenho acadêmico (IDA) quanto no indicador de ponto de virada (IPV).

Observa-se que a maior concentração de alunos com alto desempenho acadêmico ocorre em níveis de engajamento entre 7 e 10. Da mesma forma, valores mais altos de IPV estão fortemente associados a maiores níveis de engajamento.

Embora o engajamento não seja o único fator que explica o desempenho educacional, os resultados indicam que ele desempenha um papel relevante na trajetória acadêmica dos alunos, sugerindo que estratégias voltadas ao aumento do engajamento podem contribuir significativamente para a melhoria dos resultados educacionais.

Pergunta 4 - Autoavaliação (IAA)

```python
sns.scatterplot(
    data=dataset,
    x="IAA",
    y="IDA",
    alpha=0.4
)

plt.title("Autoavaliação vs Desempenho")
plt.show()

sns.scatterplot(
    data=dataset,
    x="IAA",
    y="IEG",
    alpha=0.4
)
```

A análise da relação entre autoavaliação (IAA) e os indicadores de desempenho mostra que os alunos tendem a atribuir avaliações relativamente altas a si mesmos, com grande concentração de valores entre 7 e 10. Entretanto, o desempenho acadêmico (IDA) apresenta maior variabilidade, indicando que a percepção dos alunos sobre seu próprio desempenho nem sempre corresponde exatamente aos resultados acadêmicos observados.

Em relação ao engajamento (IEG), observa-se uma associação mais consistente: alunos com maiores níveis de autoavaliação tendem também a apresentar níveis mais elevados de engajamento nas atividades. Isso sugere que a autoavaliação pode estar mais relacionada à motivação e participação dos alunos do que diretamente ao desempenho acadêmico.

Esses resultados indicam que a autopercepção positiva pode contribuir para maior envolvimento nas atividades do programa, mas não garante necessariamente melhores resultados acadêmicos, sendo necessário considerar outros fatores pedagógicos e psicossociais na análise do desempenho educacional.

Pergunta 5 - Aspectos psicossociais (IPS)

Observa-se que alunos com IPS mais baixos apresentam maior dispersão nos resultados acadêmicos e nos níveis de engajamento, incluindo casos de desempenho significativamente inferior.

Embora a relação não seja estritamente linear, os resultados sugerem que fatores psicossociais desempenham um papel importante na estabilidade da trajetória educacional dos alunos. Dessa forma, intervenções voltadas ao fortalecimento socioemocional podem contribuir para reduzir riscos de queda de desempenho ou engajamento ao longo do programa.

Da mesma forma, alguns alunos com avaliações psicopedagógicas medianas ou moderadas apresentam níveis reduzidos de defasagem.

Esses resultados sugerem que as avaliações psicopedagógicas confirmam parcialmente a defasagem identificada pelo indicador de adequação de nível (IAN), mas que outros fatores educacionais e contextuais também influenciam significativamente a trajetória acadêmica dos alunos.

Esses resultados sugerem que o engajamento nas atividades do programa pode desempenhar um papel fundamental no desenvolvimento educacional dos alunos, atuando como um dos principais fatores associados ao ponto de virada em sua trajetória acadêmica.

Pergunta 8 - Multidimensionalidade dos indicadores

```python
dataset["MEDIA_INDICADORES"] = dataset[
    ["IDA","IEG","IPS","IPP"]
].mean(axis=1)

sns.scatterplot(
    data=dataset,
    x="MEDIA_INDICADORES",
    y="INDE",
    alpha=0.4
)

plt.show()

dataset["GRUPO_INDICADORES"] = pd.qcut(
    dataset["MEDIA_INDICADORES"],
    4,
    labels=["Baixo","Médio","Alto","Muito Alto"]
)

sns.boxplot(
    data=dataset,
    x="GRUPO_INDICADORES",
    y="INDE"
)
```

A análise da combinação dos indicadores IDA, IEG, IPS e IPP mostra uma relação positiva clara com o índice global de desenvolvimento educacional (INDE). Observa-se que alunos com maiores valores na média desses indicadores apresentam níveis significativamente mais altos de INDE.

A segmentação dos alunos em grupos de desempenho ("Baixo", "Médio", "Alto" e "Muito Alto") revela um aumento progressivo do INDE conforme melhora a combinação dos indicadores. Alunos classificados no grupo "Muito Alto" apresentam as maiores médias de INDE, enquanto alunos no grupo "Baixo" concentram os menores valores.

Esses resultados indicam que o desempenho global do aluno não depende de um único fator isolado, mas sim da combinação equilibrada de múltiplas dimensões educacionais, incluindo desempenho acadêmico, engajamento nas atividades e aspectos psicossociais e psicopedagógicos.

Pergunta 9 - Previsão de risco (Machine Learning)

Foi desenvolvido um modelo preditivo utilizando Random Forest para identificar alunos em risco de defasagem educacional. O modelo foi treinado utilizando dados dos anos de 2022 e 2023 e avaliado com dados de 2024, simulando um cenário real de previsão futura.

A variável alvo foi definida considerando alunos com DEFAS ≤ -2, representando níveis moderados ou severos de defasagem educacional.

O modelo apresentou desempenho elevado, com acurácia de aproximadamente 90% e ROC AUC de 0.93, indicando excelente capacidade de identificar alunos em risco. O recall de 68% demonstra que o modelo consegue detectar a maioria dos casos de risco educacional.

A análise de importância das variáveis revelou que idade, fase educacional e o indicador de ponto de virada (IPV) são os fatores mais relevantes para prever risco de defasagem. Esses resultados indicam que o risco educacional está associado não apenas ao desempenho acadêmico, mas também à progressão educacional e ao desenvolvimento geral do aluno.

Pergunta 10 - Efetividade do programa

```python
def classificar_pedra(inde):
    
    if inde < 5.506:
        return "Quartzo"
    elif inde < 6.868:
        return "Ágata"
    elif inde < 8.230:
        return "Ametista"
    else:
        return "Topázio"

dataset["PEDRA"] = dataset["INDE"].apply(classificar_pedra)

sns.countplot(data=dataset, x="PEDRA", hue="ANO")
```

A análise da distribuição das categorias de desempenho educacional ao longo dos anos revela uma tendência positiva no desenvolvimento dos alunos participantes do programa. Observa-se crescimento nas categorias de melhor desempenho, especialmente nas categorias Ágata e Topázio, indicando aumento progressivo de alunos em níveis mais elevados de desenvolvimento educacional.

A categoria Quartzo, associada a níveis mais baixos de desempenho, apresenta redução significativa entre 2022 e 2023, sugerindo melhoria no desempenho geral dos alunos ao longo do tempo. Embora haja pequenas variações entre os anos, o crescimento consistente da categoria Topázio indica que um número cada vez maior de alunos atinge níveis elevados de desenvolvimento educacional.

Esses resultados sugerem que o programa apresenta impacto positivo no progresso educacional dos alunos, promovendo melhorias graduais nos indicadores de desempenho ao longo do ciclo de acompanhamento.
